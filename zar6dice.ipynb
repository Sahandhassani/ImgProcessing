```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
def turing_dice(imgcode):
    img1=plt.imread("d1.png",format=None)
    img2=plt.imread("d2.png",format=None)
    img3=plt.imread("d3.png",format=None)
    img4=plt.imread("d4.png",format=None)
    img5=plt.imread("d5.png",format=None)
    img6=plt.imread("d6.png",format=None)    
    img11=plt.imread("dw1.png",format=None)
    img22=plt.imread("dw2.png",format=None)
    img33=plt.imread("dw3.png",format=None)
    img44=plt.imread("dw4.png",format=None)
    img55=plt.imread("dw5.png",format=None)
    img66=plt.imread("dw6.png",format=None)
    img1=img1[0:120,0:120,:]
    img2=img2[0:120,0:120,:]
    img3=img3[0:120,0:120,:]
    img4=img4[0:120,0:120,:]
    img5=img5[0:120,0:120,:]
    img6=img6[0:120,0:120,:]
    img11=img11[0:120,0:120,:]
    img22=img22[0:120,0:120,:]
    img33=img33[0:120,0:120,:]
    img44=img44[0:120,0:120,:]
    img55=img55[0:120,0:120,:]
    img66=img66[0:120,0:120,:]
    img1=shape_change(img1,30,30)
    img2=shape_change(img2,30,30)
    img3=shape_change(img3,30,30)
    img4=shape_change(img4,30,30)
    img5=shape_change(img5,30,30)
    img6=shape_change(img6,30,30)
    img11=shape_change(img11,30,30)
    img22=shape_change(img22,30,30)
    img33=shape_change(img33,30,30)
    img44=shape_change(img44,30,30)
    img55=shape_change(img55,30,30)
    img66=shape_change(img66,30,30)

    z=img1.shape[0]
    h=z*imgcode.shape[0]
    w=z*imgcode.shape[1]
    img=np.empty((h,w,3))
    
    for i in range(imgcode.shape[0]):
        for j in range(imgcode.shape[1]):
            if imgcode[i,j,0]==1:
                img[i*z:i*z+z,j*z:j*z+z,:]=img1
            elif imgcode[i,j,0]==2:
                img[i*z:i*z+z,j*z:j*z+z,:]=img2
            elif imgcode[i,j,0]==3:
                img[i*z:i*z+z,j*z:j*z+z,:]=img3
            elif imgcode[i,j,0]==4:
                img[i*z:i*z+z,j*z:j*z+z,:]=img4
            elif imgcode[i,j,0]==5:
                img[i*z:i*z+z,j*z:j*z+z,:]=img5
            elif imgcode[i,j,0]==6:
                img[i*z:i*z+z,j*z:j*z+z,:]=img6
            elif imgcode[i,j,0]==60:
                img[i*z:i*z+z,j*z:j*z+z,:]=img66
            elif imgcode[i,j,0]==50:
                img[i*z:i*z+z,j*z:j*z+z,:]=img55
            elif imgcode[i,j,0]==40:
                img[i*z:i*z+z,j*z:j*z+z,:]=img44
            elif imgcode[i,j,0]==30:
                img[i*z:i*z+z,j*z:j*z+z,:]=img33
            elif imgcode[i,j,0]==20:
                img[i*z:i*z+z,j*z:j*z+z,:]=img22
            elif imgcode[i,j,0]==10:
                img[i*z:i*z+z,j*z:j*z+z,:]=img11                
    return img
    
def gray_(imgry):
    image=imgry.copy()
    rgb=image[:,:,0]*0.299+image[:,:,1]*0.587+image[:,:,2]*0.114
    image[:,:,0]=rgb
    image[:,:,1]=rgb
    image[:,:,2]=rgb
    image
    return image

def zar_(img3,t):
    img3[(img3[:,:,0]<t[0]) & (img3[:,:,0]>t[1])]=10
    img3[(img3[:,:,0]<t[1]) & (img3[:,:,0]>t[2])]=20
    img3[(img3[:,:,0]<t[2]) & (img3[:,:,0]>t[3])]=30
    img3[(img3[:,:,0]<t[3]) & (img3[:,:,0]>t[4])]=40
    img3[(img3[:,:,0]<t[4]) & (img3[:,:,0]>t[5])]=50
    img3[(img3[:,:,0]<t[5]) & (img3[:,:,0]>t[6])]=60
    img3[(img3[:,:,0]<t[6]) & (img3[:,:,0]>t[7])]=6
    img3[(img3[:,:,0]<t[7]) & (img3[:,:,0]>t[8])]=5
    img3[(img3[:,:,0]<t[8]) & (img3[:,:,0]>t[9])]=4
    img3[(img3[:,:,0]<t[9]) & (img3[:,:,0]>t[10])]=3
    img3[(img3[:,:,0]<t[10]) & (img3[:,:,0]>t[11])]=2
    img3[(img3[:,:,0]<t[11]) & (img3[:,:,0]>t[12])]=1

    return img3     

def shape_change(img,h,w):
    dh=round(img.shape[0]/h)
    dw=round(img.shape[1]/w)
    h=round(img.shape[0]/dh)
    w=round(img.shape[1]/dw)
    print("shape can be change to:",h,"x",w)
    img0=img[0::dh,0::dw,:]
    return img0

#load it
#img=plt.imread("profile.png",format=None)
img1=plt.imread("zibamamad.png",format=None)
img=img1[:,:,0:3]
print("img shape loaded:\n",img.shape,type(img))
print()

#size change
imgchange=shape_change(img,200,200)
print('img shape after changed',imgchange.shape)
print()

#gray it
image2=gray_(imgchange)   
plt.imshow(image2)

#1-6 pix code
tone=np.linspace(image2.max(),image2.min(),13)
print("number of dice:",imgchange.shape[0]*imgchange.shape[1])
ndice=imgchange.shape[0]*imgchange.shape[1]
print("tone:\n",tone)

#tone
img6 = zar_(image2,tone)

#turning img to dice
newimg=turing_dice(img6)
print(newimg.shape)

#show and save
plt.figure(figsize=(12,12))
plt.imshow(newimg) 
plt.imsave("zibamamad"+str(ndice)+".png",newimg)

#data frame
h=np.array(range(img6.shape[0]))
w=np.array(range(img6.shape[1]))
df6 = pd.DataFrame(img6[:,:,0],index=h,columns=w)
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)
#df6.head(500)
```
